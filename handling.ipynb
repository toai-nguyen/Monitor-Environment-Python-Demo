```python
import numpy as np
import pandas as pd
import time
import json
import keyboard
import matplotlib.pyplot as plt
```

```python
fileData = open('data.json')
data = json.load(fileData)
i = 0
```

```python
listDataTemperature = []
listDataHumidity = []
listDataWindSpeed = []
windDirectStatus = ""
x_axis = [0]
```

```python
def processData(currentData):
    currentTemp = currentData['temperature']
    currentHumidity = currentData['water']
    currentWindSpeed = currentData['windSpeed']
    currentDirect = currentData['windDirect']
    
    listDataTemperature.append(currentTemp) 
    listDataHumidity.append(currentHumidity)
    listDataWindSpeed.append(currentWindSpeed)
    
    dictData = {
        "Temperature" : {
            # "listData": listDataTemperature,
            "current": currentTemp,
            "max": max(listDataTemperature),
            "min": min(listDataTemperature),
            "avg": np.average(listDataTemperature),
        },
        "Humidity":{
            # "listData": listDataHumidity,
            "current": currentHumidity,
            "max": max(listDataHumidity),
            "min": min(listDataHumidity),
            "avg": np.average(listDataHumidity),
        },
        "WindSpeed": {
            # "listData": listDataWindSpeed,
            "current": currentWindSpeed,
            "max": max(listDataWindSpeed),
            "min": min(listDataWindSpeed),
            "avg": np.average(listDataWindSpeed),
        },
        "WindDirect": currentDirect
    }
    
    return dictData
```

```python
def drawChart(listDataHumidity, listDataTemperature, listDataWindSpeed):
        plt.ylim(0, 100)

        plt.plot(x_axis, listDataHumidity, label = 'Humidity',color = 'g')[0]
        plt.plot(x_axis, listDataTemperature, label = 'Temperature', color = 'gold')
        plt.plot(x_axis, listDataWindSpeed, label = 'Wind speed', color = 'blue')
        plt.legend()
        plt.pause(1)
        
        # graph.remove()
        plt.xlim(x_axis[0], x_axis[-1])
        x_axis.append(x_axis[-1]+1)
        
```

```python
def alert(temp, hum, speed):
    if (temp > 50 and temp < 10 and hum > 75 and speed > 50):
        return True
    else:
        return False
```

```python
def storeData(dictData):
    json_string = json.dumps(dictData)
    with open("save-data.txt", "w") as file:
      file.write(json_string)
```

```python
plt.ion()
count = 0
print("Spam Q button if u want stop")
print("Spam S button if u want save status")

while True:
  count+=1

  if(i < len(data) - 1):
    i+=1
  else:
    i = 0

  currentData = data[i]
  dictData = processData(currentData)
  
  temp = dictData["Temperature"]["current"]
  hum = dictData["Humidity"]["current"]
  speed = dictData["WindSpeed"]["current"]
  
  checkDanger = alert(temp, hum, speed)
  
  if(checkDanger):
    print("Danger time !!!")
  else:
    print("Everything is ok <3")
  
  drawChart(listDataHumidity, listDataTemperature, listDataWindSpeed)
  df = pd.DataFrame(dictData)
  print(df)
  if keyboard.is_pressed('q'):
    print('Quit the project now')
    break
  elif keyboard.is_pressed('s'):
    storeData(dictData)
    print('Save data at save-data.txt');
  if count == 10:
    print("Reaching litmit, pay more to see more !!!")
    break
  storeData(dictData)
  time.sleep(3)
```
